В качестве задания сделал наложение эффекта сепия

Се́пия (лат. sepia от др.-греч. σηπία — каракатица):
- светло-коричневое красящее вещество;
- оттенок коричневого цвета;
- цвет, присущий старым чёрно-белым фотографиям; имитация такой фотографии.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import imageio
```

```python
img = imageio.imread('cat-color.png')
print(img.shape)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img)
```

```python
def my_filter(x):
    # Sepia effect
    out = np.zeros(3, dtype=np.int32)
    out[0] = x[0] * 0.393 + x[1] * 0.769 + x[2] * 0.189 
    out[1] = x[0] * 0.349 + x[1] * 0.686 + x[2] * 0.168
    out[2] = x[0] * 0.272 + x[1] * 0.534 + x[2] * 0.131
    return out
```

```python
out_img = np.zeros(
    (img.shape[0],
     img.shape[1],
     3), 
     dtype=np.int32)
```

```python
%%time 

for i in range(out_img.shape[0]):
    for j in range(out_img.shape[1]):
        x = img[i, j]
        out_img[i, j] = my_filter(x)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(out_img)
```
